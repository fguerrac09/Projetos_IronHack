```python
import random

# lista que definem o resultado do chute/defesa
direcao = ['esquerda', 'centro', 'direita']
força = ['forte', 'fraco']
altura = ['alto', 'baixo']

#CPU escolhe a direçao
def pega_direcao():
    dire = random.choice(direcao)
    return dire

#CPU escolhe a força
def pega_força():
    forç = random.choice(força)
    return forç

#CPU escolhe a altura
def pega_altura():
    altu = random.choice(altura)
    return altu

#combinação dos 3 fatores para defesa/chute adicionando o resultado em uma lista
def pega_resultado():
    todas_defesas = []
    todas_defesas.append(pega_direcao())
    todas_defesas.append(pega_força())
    todas_defesas.append(pega_altura())
    return todas_defesas

def qual_cobrança():
    x = 1
    print(f'hora da {x}a cobrança')

#aqui o código vai pedir para o player escolher os 3 fatores do chute para o player e comparar com o resultado das escolhas da CPU e retornar se foi gol ou se o goleiro defendeu e adiconar o resultado na lista de cobranças do player.
resultado_playera = []
def pega_resultado_player():
    cpud = pega_resultado()
    
    atac_dir = input('hora da cobrança! de que lado vc quer chutar? direita, centro ou esquerda? ')
    while atac_dir != 'esquerda' and atac_dir != 'direita' and atac_dir != 'centro':
        atac_dir = input('selecione uma das 3 opcoes: direita, centro ou esquerda? ')

    atac_for = input('qual força usar? forte ou fraco? ')
    while atac_for != 'forte' and atac_for != 'fraco':
        atac_for = input('força inválida! escolher entre: forte e fraco ')

    atac_alt = input('em que altura vc vai chutar? alto ou baixo? ')
    while atac_alt != 'alto' and atac_alt != 'baixo':
        atac_alt = input('altura inválida! escolher entre: alto ou baixo ')

#adicona os resultados na lista de chute do player
    playera = []
    playera.append(atac_dir)
    playera.append(atac_for)
    playera.append(atac_alt)
    print(f'chute {player_time}: {playera}')
    print(f'defesa {cpu_time}: {cpud}')

# compara o chute do player com defesa da cpu    
    if playera == cpud:
        resultado_playera.append('defendeu')
        print('                                     DEFENDEU!!')
    elif playera[0] == cpud[0] and playera[2] == cpud[2]:
        resultado_playera.append('defendeu')  
        print('                                     DEFENDEU!!')  
    elif playera[1] == 'forte':
        r = random.random()
        if r > 0.75:
            resultado_playera.append('pra fora')  
            print('                                 PRA FORA!! O CHUTE FOI MUITO FORTE')
        else:
            resultado_playera.append('gol')
            print('                                     GOL!')
    elif playera[1] == 'fraco':
        r = random.random()
        if r > 0.75:
            resultado_playera.append('defendeu')  
            print('                                 CHUTE MUITO FRACO! O GOLEIRO SE RECUPEROU E DEFENDEU A COBRANÇA!')
        else:
            resultado_playera.append('gol')
            print('                                     GOL!')
    else:
        resultado_playera.append('gol')
        print('                                         GOL!')

#aqui o código vai pedir para o player escolher os 3 fatores da cpu e comparar com o resultado das escolhes da CPU e retornar se foi gol ou se o goleiro defendeu e adiconar o resultado na lista de cobranças do player.
resultado_playerD = []

def pega_resultado_cpu():
    cpuA = pega_resultado()
        
    atac_dir = input('hora de defender! de que lado vc quer pular? direita, centro ou esquerda? ')
    while atac_dir != 'esquerda' and atac_dir != 'direita' and atac_dir != 'centro':
        atac_dir = input('selecione uma das 3 opcoes: direita, centro ou esquerda? ')

    atac_for = input('qual força usar? forte ou fraco? ')
    while atac_for != 'forte' and atac_for != 'fraco':
        atac_for = input('força inválida! escolher entre: forte e fraco ')

    atac_alt = input('em que altura vc vai pular? alto ou baixo? ')
    while atac_alt != 'alto' and atac_alt != 'baixo':
        atac_alt = input('altura inválida! escolher entre: alto ou baixo ')

#adiciona resultado da cpuA na lista de cpuA
    playerD = []
    playerD.append(atac_dir)
    playerD.append(atac_for)
    playerD.append(atac_alt)
    print(f'defesa {player_time}: {playerD}')
    print(f'chute {cpu_time}: {cpuA}')

#compara cpuA com chute do player    
    if playerD == cpuA:
        resultado_playerD.append('defendeu')
        print('                                     DEFENDEU!!')
    elif playerD[0] == cpuA[0] and playerD[2] == cpuA[2]:
        resultado_playerD.append('defendeu')   
        print('                                     DEFENDEU!!') 
    elif cpuA[1] == 'forte':
        r = random.random()
        if r > 0.75:
            resultado_playerD.append('pra fora')  
            print('                           PRA FORA!! O CHUTE FOI MUITO FORTE')
        else:
            resultado_playerD.append('gol')
            print('                                     GOL!')
    elif cpuA[1] == 'fraco':
        r = random.random()
        if r > 0.75:
            resultado_playerD.append('defendeu')  
            print('                           CHUTE MUITO FRACO! O GOLEIRO SE RECUPEROU E DEFENDEU A COBRANÇA!')
        else:
            resultado_playerD.append('gol')
            print('                                     GOL!')    
    else:
        resultado_playerD.append('gol')
        print('                                     GOL!')

#intrudução
print('bem vindo a cobranças de penaltis!\nCada jogador tera direito a 3 cobranças e 3 tentativas de defesas. Caso haja empate o jogo será decidido na morte súbita(cobranças alternadas).')
player_time = input('escolha seu time!')
cpu_time = input('escolha o time que você quer jogar contra!')
#contador para cada time fazer 3 chutes
cont = 0
x =1
while cont < 3:
    print(f'                                {x}a cobrança')
    pega_resultado_player()
    print(f'                                {x}a tentativa de defesa')
    pega_resultado_cpu()
    cont+=1
    x+=1

gols_player = resultado_playera.count('gol')
gols_cpu =resultado_playerD.count('gol')

if gols_player < gols_cpu:
    print(f'{cpu_time} ganhou!')
elif gols_player > gols_cpu:
    print(f'{player_time} ganhou!')  

print('RESULTADO FINAL:')
print(f'{player_time}: {gols_player} gol(s)!') 
print(f'{cpu_time}: {gols_cpu} gol(s)!')


#contador da morte súbita
y = 4
while gols_cpu == gols_player:
    print('Empate! morte súbita!')
    print(f'                                {y}a cobrança!')
    pega_resultado_player()
    print(f'                                {y}a tentativa de defesa!')
    pega_resultado_cpu()
    gols_player = resultado_playera.count('gol')
    gols_cpu =resultado_playerD.count('gol')
    y+=1


    if gols_player < gols_cpu:
        print(f'{cpu_time} ganhou!')
        print('RESULTADO FINAL:')
        print(f'{player_time}: {gols_player} gol(s)!') 
        print(f'{cpu_time}: {gols_cpu} gol(s)!')
    elif gols_player > gols_cpu:
        print(f'{player_time} ganhou!') 
        print('RESULTADO FINAL:')
        print(f'{player_time}: {gols_player} gol(s)!') 
        print(f'{cpu_time}: {gols_cpu} gol(s)!') 
```
