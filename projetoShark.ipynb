```python
df_fonte=pd.read_csv('/home/felipe/Ironhack/projeto/attacks.csv', encoding = 'latin-1')
df = df_fonte.copy()
```

```python
def clean_colnames(df):
    '''Clean columns names when passing a pandas dataframe: params (df - dataframe)'''
    col_clean = []
    for col in df.columns:
        col = col.strip().lower()
        col = col.replace('.',' ')
        col_clean.append(col)
        
    df.columns = col_clean
    return df.columns
```

```python
clean_colnames(df)
```

```python
df = df.drop(axis = 1, columns = ['unnamed: 22', 'year', 'date','location', 'unnamed: 23', 'href', 'href formula', 'pdf', 'original order',
                                  'investigator or source', 'fatal (y/n)', 'case number 1', 
                                  'case number 2', 'type', 'name', 'species', 'case number', 'case number','age','sex','time','area','country'])
```

```python
df=df.dropna(subset=['activity','injury'])
```

```python
df['activity_sint'] = df['activity']
acoes = ['surfing', 'swimming', 'fishing', 'spearfishing', 'bathing', 'wading', 'diving','standing', 'snorkeling', 'scuba diving' , 'body boarding', 'body surfing', 'kayaking', 'fell overboard', 'pearl diving', 'treading water', 'boogie boarding', 'paddling', 'walking', 'cleaning fish', 'playing', 'standing', 'kite boarding' , 'spearfishing', 'feeding shark', 'bathing', 'kite surfing', 'body surfing']         
for i, row in df.iterrows():
    for acao in acoes:
        if isinstance(df['activity'][i], str) and acao in df['activity'][i].lower():
            df['activity_sint'][i]  = acao
            break
        df['activity_sint'][i] = 'others'    
         
```

```python
df['injury_sint'] = df['injury']
gravidade = ['no injury', 'minor', 'injured', 'severe', 'serious', 'fatal', 'bite','bitten','puncture','laceration', 'wounds','abrasion', 'bruise']
for i, row in df.iterrows():
    for item in gravidade:
        if isinstance(df['injury'][i], str) and item in df['injury'][i].lower():
            df['injury_sint'][i] = item
            break
        df['injury_sint'][i] = 'others'       
      
    
```

```python
df['part_sint'] = df['injury']
parte = ['leg','foot','thigh','knee','ankle','calf','heel','arm','hand','finger','toes','forearm','abdomen','face','neck']
for i, row in df.iterrows():
    for item in parte:
        if isinstance(df['injury'][i], str) and item in df['injury'][i].lower():
            df['part_sint'][i] = item
            break
        df['part_sint'][i] = 'others'    
```
